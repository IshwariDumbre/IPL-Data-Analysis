# **Dataset Name - IPL**

**Upload the dataset - 'IPL Dataset'**

**Unzip the data**

```python
import zipfile

with zipfile.ZipFile("IPL Dataset.zip", 'r') as zip_ref:
    zip_ref.extractall("ipl_dataset")
```

**Load matches and deliveries dataset**

```python
import pandas as pd

matches = pd.read_csv("ipl_dataset/matches.csv")
deliveries = pd.read_csv("ipl_dataset/deliveries.csv")
```

```python
print("Matches Dataset:")
display(matches.head())

print("\nDeliveries Dataset:")
display(deliveries.head())
```

## **"Exploratory Data Analysis (EDA)"**

(Using Matches dataset)

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

**Number of matches played per season**

```python
season_matches = matches['Season'].value_counts().sort_index()

plt.figure(figsize=(10,5))
sns.barplot(x=season_matches.index, y=season_matches.values, palette='viridis')
plt.title("Matches Played per IPL Season", fontsize=14)
plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.xticks(rotation=45)
plt.show()
```

**How many matches each team has won**

```python
team_wins = matches['winner'].value_counts()

plt.figure(figsize=(12,6))
sns.barplot(y=team_wins.index, x=team_wins.values, palette='Set2')
plt.title("Most Successful Teams in IPL (by Wins)", fontsize=14)
plt.xlabel("Number of Wins")
plt.ylabel("Teams")
plt.show()
```

**How often does the toss winner team also won the match?**

```python
toss_match_same = matches[matches['toss_winner'] == matches['winner']]
percentage = round(len(toss_match_same) / len(matches) * 100, 2)

print(f"Toss winner also won the match in {percentage}% of the matches.")
```

## **Player Performance Analysis (Using Deliveries Dataset)**

***Top 4 Run Scorers ***

```python
top_batsmen = deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(4)

plt.figure(figsize=(12,6))
sns.barplot(x=top_batsmen.values, y=top_batsmen.index, palette="crest")
plt.title("Top 4 Run Scorers in IPL", fontsize=14)
plt.xlabel("Total Runs")
plt.ylabel("Batsman")
plt.show()
```

**Most 4s and 6s**

```python
# Most 4s
fours = deliveries[deliveries['batsman_runs'] == 4]
top_fours = fours['batsman'].value_counts().head(5)

# Most 6s
sixes = deliveries[deliveries['batsman_runs'] == 6]
top_sixes = sixes['batsman'].value_counts().head(5)

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.barplot(x=top_fours.values, y=top_fours.index, palette='Reds')
plt.title("Top 5 Players with Most 4s")

plt.subplot(1,2,2)
sns.barplot(x=top_sixes.values, y=top_sixes.index, palette='Blues')
plt.title("Top 5 Players with Most 6s")

plt.tight_layout()
plt.show()
```

**Most Wickets (By Bowlers)**

```python
wicket_kinds = ['caught', 'bowled', 'lbw', 'caught and bowled', 'stumped', 'hit wicket']
wickets = deliveries[deliveries['dismissal_kind'].isin(wicket_kinds)]
top_bowlers = wickets['bowler'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_bowlers.values, y=top_bowlers.index, palette="Purples")
plt.title("Top 10 Wicket-Takers in IPL", fontsize=14)
plt.xlabel("Wickets")
plt.ylabel("Bowler")
plt.show()
```

## **Merging both Datasets- Matches and Deliveries**

**Merge matches and deliveries dataset on id and match_id**

```python
matches.columns = matches.columns.str.strip().str.lower()
deliveries.columns = deliveries.columns.str.strip().str.lower()

deliveries = deliveries.rename(columns={'match_id': 'id'})

# Merge
merged_df = deliveries.merge(matches, on='id', how='left')
merged_df.head()
```

**Teams Performance Per Season**

```python
season_wins = matches.groupby(['season', 'winner']).size().reset_index(name='wins')

top_teams = season_wins['winner'].value_counts().head(3).index

plt.figure(figsize=(12,6))
sns.barplot(data=season_wins[season_wins['winner'].isin(top_teams)],
            x='season', y='wins', hue='winner')
plt.title("Top 3 Teams: Wins Per Season")
plt.xlabel("Season")
plt.ylabel("Wins")
plt.legend(title='Team')
plt.show()
```

**Most 'Player of the Match' Winners**

```python
top_players = matches['player_of_match'].value_counts().head(7)

plt.figure(figsize=(12,6))
sns.barplot(x=top_players.values, y=top_players.index, palette='crest')
plt.title("Top 7 Player of the Match Award Winners")
plt.xlabel("Awards")
plt.ylabel("Player")
plt.show()
```

**Most Runs by Batsman and Bowlers**

```python
top_batsmen = merged_df['batsman'].value_counts().head(5).index
top_bowlers = merged_df['bowler'].value_counts().head(5).index

heatmap_data = merged_df[(merged_df['batsman'].isin(top_batsmen)) &
                         (merged_df['bowler'].isin(top_bowlers))]

pivot = heatmap_data.pivot_table(index='batsman', columns='bowler', values='batsman_runs', aggfunc='sum')

# Plot heatmap
plt.figure(figsize=(10,6))
sns.heatmap(pivot, annot=True, fmt='.0f', cmap='YlGnBu')
plt.title("Runs Scored by Batsmen vs Top Bowlers")
plt.xlabel("Bowler")
plt.ylabel("Batsman")
plt.show()
```
